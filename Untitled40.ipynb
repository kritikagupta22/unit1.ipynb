```python
import re

class Book:
    def _init_(self, title, author, isbn, price, quantity):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.price = price
        self.quantity = quantity

    def _str_(self):
        return f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}, Price: ${self.price}, Quantity: {self.quantity}"

class Bookstore:
    def _init_(self):
        self.books = []

    def add_book(self):
        print("Add Book Information:")
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        isbn = input("Enter ISBN: ")
        if not re.match("^\d{3}-\d{10}$", isbn):
            raise ValueError("Invalid ISBN format. Please use ###-########## format.")
        price = float(input("Enter price: "))
        quantity = int(input("Enter quantity: "))
        book = Book(title, author, isbn, price, quantity)
        self.books.append(book)
        print("Book added successfully!")

    def view_books(self):
        print("Book List:")
        if self.books:
            for book in self.books:
                print(book)
        else:
            print("No books in the list.")

    def search_book(self):
        criteria = input("Enter search criteria (Title / Author / ISBN): ").lower()
        if criteria not in ['title', 'author', 'isbn']:
            raise ValueError("Invalid search criteria.")
        search_query = input(f"Enter {criteria}: ").lower()

        found_books = []
        for book in self.books:
            if search_query in getattr(book, criteria).lower():
                found_books.append(book)

        if found_books:
            print("Matching Books:")
            for book in found_books:
                print(book)
        else:
            print("No books found matching the criteria.")

    def update_book(self):
        isbn = input("Enter the ISBN of the book to update: ")
        for book in self.books:
            if book.isbn == isbn:
                print("Update Book Information:")
                book.title = input("Enter updated book title: ")
                book.author = input("Enter updated author name: ")
                book.price = float(input("Enter updated price: "))
                book.quantity = int(input("Enter updated quantity: "))
                print("Book information updated successfully.")
                return
        print("Book not found.")

    def delete_book(self):
        isbn = input("Enter the ISBN of the book to delete: ")
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print("Book deleted successfully.")
                return
        print("Book not found.")

    def calculate_total_revenue(self):
        total_revenue = sum(book.price * book.quantity for book in self.books)
        print(f"Total Revenue: ${total_revenue:.2f}")

    def display_out_of_stock(self):
        out_of_stock_books = [book for book in self.books if book.quantity == 0]
        if out_of_stock_books:
            print("Out of Stock Books:")
            for book in out_of_stock_books:
                print(book)
        else:
            print("No out of stock books.")

    def sort_books_by_price(self):
        sorted_books = sorted(self.books, key=lambda x: x.price)
        print("Books Sorted by Price:")
        for book in sorted_books:
            print(book)

    def sell_book(self):
        isbn = input("Enter the ISBN of the book to sell: ")
        for book in self.books:
            if book.isbn == isbn:
                quantity_sold = int(input(f"Enter the quantity of '{book.title}' to sell: "))
                if quantity_sold <= book.quantity:
                    book.quantity -= quantity_sold
                    print(f"{quantity_sold} copy(s) of '{book.title}' sold successfully.")
                    return
                else:
                    print("Insufficient stock available.")
                    return
        print("Book not found.")

    def display_menu(self):
        print("\n===== Bookstore Management System =====")
        print("1. Add Book")
        print("2. View Books")
        print("3. Search Book")
        print("4. Update Book Information")
        print("5. Delete Book")
        print("6. Calculate Total Revenue")
        print("7. Display Out of Stock Books")
        print("8. Sort Books by Price")
        print("9. Sell Book")
        print("10. Exit")

    def run(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")

            try:
                if choice == '1':
                    self.add_book()
                elif choice == '2':
                    self.view_books()
                elif choice == '3':
                    self.search_book()
                elif choice == '4':
                    self.update_book()
                elif choice == '5':
                    self.delete_book()
                elif choice == '6':
                    self.calculate_total_revenue()
                elif choice == '7':
                    self.display_out_of_stock()
                elif choice == '8':
                    self.sort_books_by_price()
                elif choice == '9':
                    self.sell_book()
                elif choice == '10':
                    print("Thank you for using the Bookstore Management System.")
                    break
                else:
                    print("Invalid choice. Please enter a valid option.")
            except Exception as e:
                print(f"Error: {e}")

    bookstore = Bookstore()
    bookstore.run()
```
