```python
#Write a program to check if whether is even as well as prime{as all prime numbers are odd expect two}.

def prime():
  num1=int(input("enter the number to check whether the number is prime or not"))
  for i in range(num1):
    if(i<=1):
      continue
    elif(num1%i==0):
        print("the given number is not prime")
        break
    else:
          continue
  if(i+1==num1):
    print("the given number is prime")
prime()
```

```python
def fact(n):
    if(n==0 or n==1):
       return 1
    else:
       return n*fact(n-1)
5*fact(4)
4*fact(3)
3*fact(2)
2*fact(1)
```
