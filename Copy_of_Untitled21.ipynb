```python
import re

books = []

def add_book():
    print("Add Book Information:")
    title = input("Enter book title: ")
    author = input("Enter author name: ")
    isbn = input("Enter ISBN: ")
    if not re.match("^\d{3}-\d{10}$", isbn):
        print("Invalid ISBN format. Please use ###-########## format.")
        return
    price = float(input("Enter price: "))
    books.append({"title": title, "author": author, "isbn": isbn, "price": price})
    print("Book added successfully!")

def view_books():
    print("Book List:")
    if books:
        for book in books:
            print(f"Title: {book['title']}, Author: {book['author']}, ISBN: {book['isbn']}, Price: ${book['price']}")
    else:
        print("No books in the list.")

def search_book():
    criteria = input("Enter search criteria (Title / Author / ISBN): ").lower()
    search_query = input(f"Enter {criteria}: ").lower()

    found_books = []
    for book in books:
        if search_query in book[criteria].lower():
            found_books.append(book)

    if found_books:
        print("Matching Books:")
        for book in found_books:
            print(f"Title: {book['title']}, Author: {book['author']}, ISBN: {book['isbn']}, Price: ${book['price']}")
    else:
        print("No books found matching the criteria.")

def update_book():
    isbn = input("Enter ISBN of the book to update: ")
    for book in books:
        if book['isbn'] == isbn:
            print("Update Book Information:")
            book['title'] = input("Enter updated title: ")
            book['author'] = input("Enter updated author name: ")
            book['price'] = float(input("Enter updated price: "))
            print("Book information updated successfully.")
            return
    print("Book not found.")

def delete_book():
    isbn = input("Enter ISBN of the book to delete: ")
    for book in books:
        if book['isbn'] == isbn:
            books.remove(book)
            print("Book deleted successfully.")
            return
    print("Book not found.")

def calculate_total_revenue():
    total_revenue = sum(book['price'] for book in books)
    print(f"Total revenue: ${total_revenue:.2f}")

def display_top_selling_books():
    if books:
        sorted_books = sorted(books, key=lambda x: x['price'], reverse=True)
        print("Top Selling Books:")
        for i, book in enumerate(sorted_books[:5]):
            print(f"{i+1}. Title: {book['title']}, Author: {book['author']}, Price: ${book['price']}")
    else:
        print("No books in the list.")

def main_menu():
    print("\n===== Bookstore Management System =====")
    print("1. Add Book")
    print("2. View Books")
    print("3. Search Book")
    print("4. Update Book")
    print("5. Delete Book")
    print("6. Calculate Total Revenue")
    print("7. Display Top Selling Books")
    print("8. Exit")

def main():
    while True:
        main_menu()
        choice = input("Enter your choice: ")

        if choice == '1':
            add_book()
        elif choice == '2':
            view_books()
        elif choice == '3':
            search_book()
        elif choice == '4':
            update_book()
        elif choice == '5':
            delete_book()
        elif choice == '6':
            calculate_total_revenue()
        elif choice == '7':
            display_top_selling_books()
        elif choice == '8':
            print("Thank you for using the Bookstore Management System.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
